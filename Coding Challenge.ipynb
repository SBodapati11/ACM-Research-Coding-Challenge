```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as graph
import numpy as np
from sklearn.mixture import GaussianMixture

dataset = pd.read_csv('ClusterPlot.csv')
dataset.drop('Unnamed: 0', axis = 1, inplace = True)
print(dataset.head())
```

```python
x_values = dataset['V1']
y_values = dataset['V2']
graph.grid(zorder=0, color="lightgrey")
graph.scatter(x_values, y_values, zorder=3)
```

```python
num_array = np.arange(1,15)
model_list = []
for num in num_array:
    model = GaussianMixture(num, covariance_type='full', random_state=0).fit(dataset)
    model_list.append(model)

aic_list = []
bic_list = []
for model in model_list:
    aic_list.append(model.aic(dataset))
    bic_list.append(model.bic(dataset))
    
graph.plot(num_array, aic_list, color='red')
graph.plot(num_array, bic_list, color='blue')
```

```python
model1 = GaussianMixture(n_components=2).fit(dataset)
label1 = model1.predict(dataset)
graph.scatter(x_values, y_values, c=label1)
```

```python
model2 = GaussianMixture(n_components=3).fit(dataset)
label2 = model2.predict(dataset)
graph.scatter(x_values, y_values, c=label2)
```

```python
prob_list_1 = model1.predict_proba(dataset)
print(prob_list_1)
```

```python
prob_list_2 = model2.predict_proba(dataset)
print(prob_list_2)
```
